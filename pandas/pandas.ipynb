### Задание 1 

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
movie_data = pd.read_csv('ml-latest-small/movies.csv')
```

```python
movie_data.head()
```

```python
scores_data = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
scores_data.head()
```

```python
top_rated_movies = scores_data[scores_data.rating == 5] # отбор фильмов с оценкой 5
```

```python
top_movies_sort = top_rated_movies['movieId'].value_counts().index.tolist() # подсчет кол-ва оценок по каждому фильму, id фильма в индексе

if top_movies_sort:
    top_rated_movie = top_movies_sort[0] # если нашлись фильмы с оценкой 5, то берется наиболее часто встречающийся
    print(movie_data[movie_data.movieId == top_rated_movie].title)
else:
    print('нет фильмов с оценкой 5')
```

### Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_data = pd.read_csv('power.csv')
```

```python
power_data.head()
```

```python
Baltic_countries = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]
```

```python
Baltic_power_data = power_data[(power_data.country.isin(Baltic_countries)) \
                   & (power_data.year >= 2005) \
                   & (power_data.year <= 2010) \
                   & (power_data.quantity >= 0) \
                   & (power_data.category.isin(categories))]
```

```python
Baltic_power_data['quantity'].sum()
```
