## Решение

### Исходные данные

````python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
````

Пользователь по команде "*p*" может узнать владельца документа по его номеру

````python
def find_person_by_document(doc_number, documents):
    '''
    Поиск владельца документа по номеру
    
    doc_number - номер документа
    documents - данные документов
    '''
    person = None
    for document in documents:
        if doc_number == document['number']:
            person = document['name']
            break
            
    if person is not None:
        return f'Владелец документа: {person}'
    else:
        return 'Документ не найден в базе'
````

Пользователь по команде "*s*" может по номеру документа узнать на какой полке он хранится

````python
def find_shelf_by_document(doc_number, directories):
    '''
    Поиск полки, на которой лежит документ по номеру документа
    
    doc_number - номер документа
    directories - данные документов на полках
    '''
    shelf = None
    for directory in directories:
        for document in directories[directory]:
            if doc_number == document:
                shelf = directory
                break
            
    if shelf is not None:
        return (f'Документ хранится на полке: {shelf}', shelf)
    else:
        return ('Документ не найден в базе', shelf)
````

Пользователь по команде "*l*" может увидеть полную информацию по всем документам

````python
def show_documents(documents, directories):
    '''
    Показать информацию по всем документам
    
    documents - список документов
    '''
    result = ''
    for document in documents:
        _, shelf = find_shelf_by_document(document["number"], directories)
        result += f'№: {document["number"]}, тип: {document["type"]}, владелец: {document["name"]}, полка хранения: {shelf}\n'

    return result
````

Пункт 4. Пользователь по команде "*ads*" может добавить новую полку

````python
def add_shelf(shelf_number, directories):
    '''
    Добавить полку по указанному номеру
    
    shelf_number - номер добавляемой полки
    directories - данные полок
    '''
    shelf_list = list(directories.keys())
    for shelf in directories:
        if shelf_number == shelf:
            return f'Такая полка уже существует. Текущий перечень полок: {",".join(shelf_list)}'
   
    shelf_list.append(shelf_number) 
    directories[shelf_number] = []
    return f'Полка добавлена. Текущий перечень полок:{",".join(shelf_list)}'
````

Пользователь по команде "*ds*" может удалить существующую полку из данных (только если она пустая)

````python
def delete_shelf(shelf_number, directories):
    '''
    Удалить полку по указанному номеру (только если она пустая)
    
    shelf_number - номер удаляемой полки
    directories - данные полок
    '''
    shelf_list = list(directories.keys())
    for shelf in directories:
        if shelf_number == shelf:
            if len(directories[shelf]) == 0:
                del directories[shelf]
                return f'Полка удалена. Текущий перечень полок: {",".join(list(directories.keys()))}'
            else:
                return f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {",".join(shelf_list)}'
            
    return f'Такой полки не существует. Текущий перечень полок: {",".join(shelf_list)}'
````

Пользователь по команде "*ad*" может добавить новый документ в данные

````python
def add_document(doc_number, person, doc_type, shelf, documents, directories):
    '''
    Добавить документ
    
    doc_number - номер документа
    person - имя владельца документа
    doc_type - тип документа
    shelf - полка, на которую нужно положить документ
    documents - данные документов
    directories - данные полок
    '''
    if shelf not in directories.keys():
        return 'Такой полки не существует. Добавьте полку командой as.\nТекущий список документов:\n' + show_documents(documents, directories)
    else:
        directories[shelf].append(doc_number)
        doc_data = {'type': doc_type, 'number': doc_number, 'name': person}
        documents.append(doc_data)
        return 'Документ добавлен. Текущий список документов:\n' + show_documents(documents, directories)
````

Пользователь по команде "*d*" может удалить документ из данных

````python
def delete_document_from_directiory(doc_number, directories):
    '''
    Удалить документ с полки по номеру
    
    doc_number - номер документа
    directories - данные полок
    '''
    for shelf_data in directories.values():
        if doc_number in shelf_data:
            shelf_data.remove(doc_number)
            return
````

````python
def delete_document(doc_number, documents, directories):
    '''
    Удалить документ
    
    doc_number - номер документа
    documents - данные документов
    directories - данные полок
    '''
    for document in documents:
        if doc_number == document['number']:
            delete_document_from_directiory(doc_number, directories) #удаление документа с полки
            documents.remove(document) #удаление документа из документов
            return 'Документ удален.\nТекущий список документов:\n' + show_documents(documents, directories)
        
    return 'Документ не найден в базе.\nТекущий список документов:\n' + show_documents(documents, directories)
````

Пользователь по команде "*m*" может переместить документ с полки на полку

````python
def move_document(doc_number, shelf, directories):
    '''
    Переместить документ на другую полку
    
    doc_number - номер документа
    shelf - номер полки, на которую нужно переместить документ
    directories - данные полок
    '''
    
    _, old_shelf = find_shelf_by_document(doc_number, directories)
    if old_shelf is None:
        return 'Документ не найден в базе.\nТекущий список документов:\n' + show_documents(documents, directories)
    elif shelf not in directories.keys():
        return f'Такой полки не существует. Текущий перечень полок: {",".join(list(directories.keys()))}'
    else:
        directories[old_shelf].remove(doc_number)
        directories[shelf].append(doc_number)
        return 'Документ перемещен.\nТекущий список документов:\n' + show_documents(documents, directories)
    
````

### Запуск программы

````python
def main(documents, directories):
    print('''
    Список доступных команд:
    p - узнать владельца документа по номеру
    s - найти полку, на которой лежит документ по его номеру
    l - вывести список документов
    ads - добавить полку
    ds - удалить полку
    ad - добавить документ
    d - удалить документ
    m - переместить документ
    q - выйти из программы 
    ''')
    while True:
        command = input('Введите команду: ')
        if command == 'p':
            doc_number = input('Введите номер документа: ')
            print(find_person_by_document(doc_number, documents))
        elif command == 's':
            doc_number = input('Введите номер документа: ')
            print(find_shelf_by_document(doc_number, directories)[0])
        elif command == 'l':
            print(show_documents(documents, directories))
        elif command == 'ads':
            shelf_number = input('Введите номер полки: ')
            print(add_shelf(shelf_number, directories))
        elif command == 'ds':
            shelf_number = input('Введите номер полки: ')
            print(delete_shelf(shelf_number, directories))
        elif command == 'ad':
            doc_number = input('Введите номер документа: ')
            doc_type = input('Введите тип документа: ')
            person = input('Введите владельца документа: ')
            shelf_number = input('Введите номер полки: ')
            print(add_document(doc_number, person, doc_type, shelf_number, documents, directories))
        elif command == 'd':
            doc_number = input('Введите номер документа: ')
            print(delete_document(doc_number, documents, directories))
        elif command == 'm':
            doc_number = input('Введите номер документа: ')
            shelf_number = input('Введите номер полки: ')
            print(move_document(doc_number, shelf_number, directories))
        elif command == 'q':
            print('До свидания')
            break
        else:
            print('Неизвестная команда')
````

````python
main(documents, directories)
````
